```
# chenge type to float and errors='coerce' any error to NaN
df['BasePay']=pd.to_numeric(df['BasePay'],errors='coerce')
df['OvertimePay']=pd.to_numeric(df['OvertimePay'],errors='coerce')
df['OtherPay']=pd.to_numeric(df['OtherPay'],errors='coerce')
df['Benefits']=pd.to_numeric(df['Benefits'],errors='coerce')
```

EDA--> Exploratory Dat Analysis

1- what is average basepsy?

```
avg_basepay= df['BasePay'].mean()
print(f'the average basepay is: {avg_basepay}')
```

2- what is the lowest basepsy?

```
min_basepay= df['BasePay'].min()
print(f'the lowest basepay is: {min_basepay}')
```

3- what is the highest basepsy?

```
hig_basepay= df['BasePay'].max()
print(f'the highest basepay is: {hig_basepay}')
```

4- who takes the highest basepsy?

```
df[df['BasePay']==hig_basepay]['EmployeeName']
```

5-what is the average basepsy for each job title?

```
# ues gropbey  بتجمع المتشابهات  وبيتبعها على طول عمود نص
result =  df.groupby('JobTitle')['BasePay'].mean()
result.sort_values(ascending=False).head()
```
